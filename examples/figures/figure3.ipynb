```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from os.path import join

sns.set(context='poster', style='ticks')
```

```python
data_dir = '/Users/kbottenh/Dropbox/Projects/physics-retrieval/data'
fig_dir = '/Users/kbottenh/Dropbox/Projects/physics-retrieval/figures'
```

```python
df = pd.read_csv(join(data_dir, 
                      'rescored', 
                      'physics_learning-nonbrain_OLS-BayesianImpute.csv'), 
                 header=0, index_col=0)
```

```python
outcomes = ['Phy48Grade', 'RetrPhyAcc2', 'FCIPhyAcc2', 
            'deltaRetrPhyAcc', 'deltaFCIPhyAcc']
covariates = ['Age', 'Mod', 'Strt.Level', 'F']
```

```python
crayons_l = sns.crayon_palette(['Vivid Tangerine', 'Cornflower'])
crayons_m = sns.crayon_palette(['Red Orange', 'Cerulean'])
crayons_d = sns.crayon_palette(['Brick Red', 'Midnight Blue'])
grays = sns.light_palette('#999999', n_colors=3, reverse=True)

f_2 = sns.crayon_palette(['Red Orange', 'Vivid Tangerine'])
m_2 = sns.crayon_palette(['Cerulean', 'Cornflower'])

sns.set_palette(grays)
```

```python
iqs = ['VCI', 'WMI', 'PSI', 'PRI', 'FSIQ']
for iq in iqs:
    df['delta{0}'.format(iq)] = df['{0}2'.format(iq)] - df['{0}1'.format(iq)]
    df['delta{0}XSex'.format(iq)] = df['delta{0}'.format(iq)] * df['F']
    df['{0}2XSex'.format(iq)] = df['{0}2'.format(iq)] * df['F']
    df['delta{0}XClass'.format(iq)] = df['delta{0}'.format(iq)] * df['Mod']
    df['{0}2XClass'.format(iq)] = df['{0}2'.format(iq)] * df['Mod']
    df['SexXClass'] = df['F'] * df['Mod']
    df['delta{0}XSexXClass'.format(iq)] = df['delta{0}'.format(iq)] * df['SexXClass']
    df['{0}2XSexXClass'.format(iq)] = df['{0}2'.format(iq)] * df['SexXClass']
```

```python
plt.tight_layout(pad=5)
g = sns.lmplot('deltaPRI', 'FCIPhyAcc2', data=df, hue='F', hue_order=[1.0,0.0],
           palette=crayons_d, legend=False, height=7., )
#g.set_yticklabels(['10', '7.5', '5.0', '2.5', '0.0', '-2.5', '-5.0', '-7.5', '-10'])
g.set_xlabels('Change in perceptual reasoning')
g.set_ylabels('Physics reasoning accuracy')
g.savefig(join(fig_dir, 'deltaPRIXsex-fci2_acc-imp.png'), dpi=300)
```
