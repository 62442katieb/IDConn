```python
import numpy as np
import pandas as pd
import seaborn as sns

from os.path import join
from statsmodels.stats.anova import anova_lm
import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt

from sklearn.impute import KNNImputer
sns.set_context('paper')
```

```python
def liji_sidak_mc(data, alpha):
    import math
    import numpy as np

    mc_corrmat = data.corr()
    mc_corrmat.fillna(0, inplace=True)
    eigvals, eigvecs = np.linalg.eig(mc_corrmat)

    M_eff = 0
    for eigval in eigvals:
        if abs(eigval) >= 0:
            if abs(eigval) >= 1:
                M_eff += 1
            else:
                M_eff += abs(eigval) - math.floor(abs(eigval))
        else:
            M_eff += 0
    print('Number of effective comparisons: {0}'.format(M_eff))

    #and now applying M_eff to the Sidak procedure
    sidak_p = 1 - (1 - alpha)**(1/M_eff)
    if sidak_p < 0.00001:
        print('Critical value of {:.3f}'.format(alpha),'becomes {:2e} after corrections'.format(sidak_p))
    else:
        print('Critical value of {:.3f}'.format(alpha),'becomes {:.6f} after corrections'.format(sidak_p))
    return sidak_p, M_eff
```

```python
sink_dir = '/Users/kbottenh/Dropbox/Projects/physics-retrieval/data/output'
data_dir = '/Users/kbottenh/Dropbox/Projects/physics-retrieval/data'
fig_dir = '/Users/kbottenh/Dropbox/Projects/physics-retrieval/figures/'

masks = ['shen2015', 'craddock2012']

tasks = {'retr': [{'conditions': ['Physics', 'General']},
                  {'runs': [0,1]}], 
         'fci': [{'conditions': ['Physics', 'NonPhysics']},
                  {'runs': [0,1,2]}]}

sessions = [0,1]
sesh = ['pre', 'post']
conds = ['high-level', 'lower-level']

post_iq = ['VCI2', 'PRI2', 'WMI2', 'PSI2', 'FSIQ2']
delta_iq = ['deltaVCI', 'deltaPRI', 'deltaWMI', 'deltaPSI', 'deltaFSIQ']
iq_vars = post_iq + delta_iq
iqs = ['VCI', 'PRI', 'WMI', 'PSI', 'FSIQ']

outcomes = ['Phy48Grade', 'RetrPhyAcc2', 'FCIPhyAcc2']
covariates = ['Age', 'Mod', 'StrtLvl', 'F']
```

```python
all_df = pd.read_csv(join(data_dir, 'rescored', 'physics_learning-nonbrain.csv'), index_col=0)
```

```python
ols_df = all_df.drop(['Ethnic.Grp', 'Handedness', 'deltaFCIPhyAcc',
                      'Mean Correct RT Pre', 'Mean Correct RT Post', 'deltaRetrPhyAcc',
                      'deltaVCI', 'deltaPRI', 'deltaWMI', 'deltaPSI', 'deltaFSIQ', 'GPA'], axis=1)
```

```python
dumb = pd.get_dummies(ols_df['Sex'])
ols_df['F'] = dumb['F']
dumb = pd.get_dummies(ols_df['Class.Type'])
ols_df['Mod'] = dumb['Mod']
ols_df['StrtLvl'] = ols_df['Strt.Level'] / 10
ols_df.drop(['Sex', 'Class.Type', 'Strt.Level'], inplace=True, axis=1)
```

```python
#compute deltas
for iq in iqs:
    imp_df['delta{0}'.format(iq)] = imp_df['{0}2'.format(iq)] - imp_df['{0}1'.format(iq)]
```

```python
adj_a, _ = liji_sidak_mc(imp_df[outcomes + covariates + iq_vars], 0.01)
```

```python
simple = {}
intrxn = {}
compar = {}
for outcome in outcomes:
    for iq_meas in iq_vars:
        reg_form = '{0} ~ {1} + F + Mod + F*Mod + Age + StrtLvl'.format(outcome, iq_meas)
        mod = smf.ols(formula=reg_form, data=imp_df)
        res = mod.fit()
        simple['{0}~{1}'.format(outcome, iq_meas)] = res
        
        if res.get_robustcov_results().f_pvalue <= adj_a:
            print('\n*{0} on {1} simple model IS significant'.format(outcome, iq_meas))
            if res.get_robustcov_results().pvalues[1] < 0.05:
                print('* and {0} coefficient significant, too!'.format(iq_meas))
            else:
                print('x but {0} coefficient not significant.'.format(iq_meas))
        else:
            print('\n{0} on {1} simple model NOT significant'.format(outcome, iq_meas))
        
        itx_form = '{0} ~ {1} + {1}*F + {1}*Mod + {1}*F*Mod + F + Mod + F*Mod + Age + StrtLvl'.format(outcome, 
                                                                                                       iq_meas)   
        mod = smf.ols(formula=itx_form, data=imp_df)
        res2 = mod.fit()
        intrxn['{0}~{1}'.format(outcome, iq_meas)] = res2
        
        if res2.get_robustcov_results().f_pvalue <= adj_a:
            print('*{0} on {1} interaction model IS significant'.format(outcome, iq_meas))
            if res2.get_robustcov_results().pvalues[1] < 0.05:
                print('* and {0} coefficient significant, too!'.format(iq_meas))
            else:
                print('x but {0} coefficient not significant.'.format(iq_meas))
        else:
            print('{0} on {1} interaction model NOT significant'.format(outcome, iq_meas))
        
        compare = anova_lm(res.get_robustcov_results(), res2.get_robustcov_results())
        compar['{0}~{1}'.format(outcome, iq_meas)] = compare
        if compare['Pr(>F)'][1] <= 0.05:
            print(np.round(compare['Pr(>F)'][1], 3), 
                  'interaction model fits better than nested model for {0} ~ {1}'.format(outcome, iq_meas))
        elif compare['Pr(>F)'][0] <= 0.05:
            print('Nested model fits better than interaction model for {0} ~ {1}'.format(outcome, iq_meas))
        else:
            print(np.round(compare['Pr(>F)'][1], 3), 'no difference between models')
```
